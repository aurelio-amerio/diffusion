```python
from scoresbibm.utils.data_utils import query, get_summary_df, load_model
from scoresbibm.utils.plot import plot_metric_by_num_simulations, use_style,multi_plot

import matplotlib.pyplot as plt

import logging
logging.getLogger('matplotlib.font_manager').disabled = True
```

```python
DEFAULT_COLORS = {"npe": "#154c79", "nle": "#1e81b0", "nre": "#76b5c5", "nspe": "#fdae61", "score_transformer": "#d73027", "score_transformer_posterior": "#8b27d7", "score_transformer_directed": "#f25d1d","score_transformer_min_graphical": "#f25d1d", "score_transformer_undirected": "#911c1c", "score_transformer_graphical": "#911c1c"}
figsize_per_row = 1.55
figsize_per_col = 1.85
```

```python
with use_style("pyloric"):
    fig, axes = multi_plot("../../results/bm_main", "metric", "task", plot_metric_by_num_simulations, hue="method", color_map=DEFAULT_COLORS, figsize_per_row=1.05, figsize_per_col=2.5, legend_ncol=4, method=["npe", "score_transformer", "score_transformer_undirected", "score_transformer_directed"],method_sde_name="vesde", alpha=0.8)
    legend = fig.legends[-1]
    axes[0,0].set_title("C2ST posterior")
    axes[0,0].set_ylabel("Lin. Gaussian")
    axes[1,0].set_ylabel("Mix. Gaussian")
    for i in range(4):
        axes[i,0].set_yticks([0.5,1.0])
    fig.legends = []
    fig.savefig("sbibm_task_reduced_methods.svg")
    plt.show()
```

```python
with use_style("pyloric"):
    fig = plt.figure(figsize=(3,0.5))
    ax = plt.gca()
    plt.axis("off")
    fig.legend(ncol=4, handles=[legend.legendHandles[3], legend.legendHandles[0],legend.legendHandles[2],legend.legendHandles[1]], labels=[ "NPE","Simformer",  "    Simformer\n (undirected graph)","    Simformer\n (directed graph)"], handletextpad=0.5,borderaxespad=0,loc="center", bbox_to_anchor=(-0.1, 0.5))
    fig.savefig("method_legend_reduced.svg", transparent=True, bbox_inches="tight")
    plt.show()
```

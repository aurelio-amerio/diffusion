```python
import svgutils.transform as sg
import sys 

#create new SVG figure
fig = sg.SVGFigure(width="10cm", height="9cm")

# load matpotlib-generated figures
fig1 = sg.fromfile('sbibm_task_reduced_methods.svg')
fig2 = sg.fromfile('all_cond_task_reduced_methods.svg')
fig12_legend = sg.fromfile('method_legend_reduced.svg')

# get the plot objects
plot1 = fig1.getroot()
plot2 = fig2.getroot()
plot3 = fig12_legend.getroot()
plot1.moveto(0, 10)
plot2.moveto(200, 10)
plot3.moveto(5, 300)


# add text labels
txt1 = sg.TextElement(5,15, "a", size=12, weight="bold", font="Arial")
txt2 = sg.TextElement(205,15, "b", size=12, weight="bold", font="Arial")

# append plots and labels to figure
fig.append([plot1, plot2,plot3])
fig.append([txt1, txt2])

fig.save("fig4.svg")
```

```python
from IPython.display import SVG

SVG(fig.to_str())
```
